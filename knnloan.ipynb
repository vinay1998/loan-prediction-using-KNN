```python
import pandas
import math
```

```python
df=pandas.read_csv("train.csv")
df_test=pandas.read_csv("test.csv")
df_check=pandas.read_csv("check.csv")
```

```python
#preprocessing



# preprocessing dependents coloumn
zero=one=two=three=0
for i in range(0,len(df)):
    if df.loc[i,"Dependents"]=="3+":
        df.loc[i,"Dependents"]=3
        three=three+1
    elif df.loc[i,"Dependents"]=="0":
        zero=zero+1
    elif df.loc[i,"Dependents"]=="1":
        one=one+1
    elif df.loc[i,"Dependents"]=="2":
        two=two+1
frequent_count=[zero,one,two,three]
frequent_item=max(frequent_count)
if frequent_item==zero:
    replace=0
elif frequent_item==one:
    replace=1
elif frequent_item==two:
    replace=2
elif frequent_item==three:
    replace=3
for i in range(0,len(df)):
    if math.isnan(float(df.loc[i,"Dependents"])):
        df.loc[i,"Dependents"]=replace

#preprocessing education coloumn
graduate=not_graduate=0
for i in range(0,len(df)):
    if df.loc[i,"Education"]=="Graduate":
        df.loc[i,"Education"]=1
        graduate=graduate+1
    elif df.loc[i,"Education"]=="Not Graduate":
        df.loc[i,"Education"]=2
        not_graduate=not_graduate+1
frequent_count=[graduate,not_graduate]
frequent_item=max(frequent_count)
if frequent_item==graduate:
    replace=1
elif frequent_item==not_graduate:
    replace=2
for i in range(0,len(df)):
    if math.isnan(float(df.loc[i,"Education"])):
        df.loc[i,"Education"]=replace

#preprocessing self_employed coloumn
self=not_self=0
for i in range(0,len(df)):
    if df.loc[i,"Self_Employed"]=="Yes":
        df.loc[i,"Self_Employed"]=2
        self=self+1
    elif df.loc[i,"Self_Employed"]=="No":
        df.loc[i,"Self_Employed"]=1
        not_self=not_self+1
frequent_count=[self,not_self]
frequent_item=max(frequent_count)
if frequent_item==self:
    replace=2
elif frequent_item==not_self:
    replace=1
for i in range(0,len(df)):
    if math.isnan(float(df.loc[i,"Self_Employed"])):
        df.loc[i,"Self_Employed"]=replace
 
"""preprocessing ApplicantIncome"""
numerator=denominator=0
for i in range(0,len(df)):
    if not(math.isnan(float(df.loc[i,"ApplicantIncome"]))):
        numerator=numerator+float(df.loc[i,"ApplicantIncome"])
        denominator=denominator+1
replace=mean=numerator/denominator
for i in range(0,len(df)):
    if math.isnan(float(df.loc[i,"ApplicantIncome"])):
        df.loc[i,"ApplicantIncome"]=int(replace)
        
#preprocessing CoapplicantIncome coloumn
numerator=denominator=0
for i in range(0,len(df)):
    if not(math.isnan(float(df.loc[i,"CoapplicantIncome"]))):
        numerator=numerator+float(df.loc[i,"CoapplicantIncome"])
        denominator=denominator+1
replace=mean=numerator/denominator
for i in range(0,len(df)):
    if math.isnan(float(df.loc[i,"CoapplicantIncome"])):
        df.loc[i,"CoapplicantIncome"]=int(replace)
        
#preprocessing LoanAmount coloumn
numerator=denominator=0
for i in range(0,len(df)):
    if not(math.isnan(float(df.loc[i,"LoanAmount"]))):
        numerator=numerator+float(df.loc[i,"LoanAmount"])
        denominator=denominator+1
replace=mean=numerator/denominator
for i in range(0,len(df)):
    if math.isnan(float(df.loc[i,"LoanAmount"])):
        df.loc[i,"LoanAmount"]=int(replace)

#preprocessing Loan_Amount_Term   given values in months
numerator=denominator=0
for i in range(0,len(df)):
    if not(math.isnan(float(df.loc[i,"Loan_Amount_Term"]))):
        numerator=numerator+float(df.loc[i,"Loan_Amount_Term"])
        denominator=denominator+1
replace=mean=numerator/denominator
for i in range(0,len(df)):
    if math.isnan(float(df.loc[i,"Loan_Amount_Term"])):
        df.loc[i,"Loan_Amount_Term"]=int(replace)
     
#preprocessing Credit_History coloumn
numerator=denominator=0
for i in range(0,len(df)):
    if not(math.isnan(float(df.loc[i,"Credit_History"]))):
        numerator=numerator+float(df.loc[i,"Credit_History"])
        denominator=denominator+1
replace=mean=numerator/denominator
for i in range(0,len(df)):
    if math.isnan(float(df.loc[i,"Credit_History"])):
        df.loc[i,"Credit_History"]=int(replace)

#preprocessing Property_Area coloumn
urban=semi_urban=rural=0
for i in range(0,len(df)):
    if df.loc[i,"Property_Area"]=="Urban":
        df.loc[i,"Property_Area"]=1
        urban=urban+1
    elif df.loc[i,"Property_Area"]=="Semiurban":
        df.loc[i,"Property_Area"]=2
        semi_urban=semi_urban+1
    elif df.loc[i,"Property_Area"]=="Rural":
        df.loc[i,"Property_Area"]=3
        rural=rural+1
frequent_count=[urban,semi_urban,rural]
frequent_item=max(frequent_count)
if frequent_item==urban:
    replace=1
elif frequent_item==semi_urban:
    replace=2
elif frequent_item==rural:
    replace=3
for i in range(0,len(df)):
    if math.isnan(int(df.loc[i,"Property_Area"])):
        df.loc[i,"Property_Area"]=replace
        
#preprocessing Loan_Status coloumn
yes=no=0
for i in range(0,len(df)):
    if df.loc[i,"Loan_Status"]=="Y":
        df.loc[i,"Loan_Status"]=1
        yes=yes+1
    elif df.loc[i,"Loan_Status"]=="N":
        df.loc[i,"Loan_Status"]=2
        no=no+1
frequent_count=[yes,no]
frequent_item=max(frequent_count)
if frequent_item==yes:
    replace=1
elif frequent_item==no:
    replace=2
for i in range(0,len(df)):
    if math.isnan(int(df.loc[i,"Loan_Status"])):
        df.loc[i,"Loan_Status"]=replace
        
#preprocessing Loan_Status coloumn for check
yes=no=0
for i in range(0,len(df_check)):
    if df_check.loc[i,"Loan_Status"]=="Y":
        df_check.loc[i,"Loan_Status"]=1
        yes=yes+1
    elif df_check.loc[i,"Loan_Status"]=="N":
        df_check.loc[i,"Loan_Status"]=2
        no=no+1
frequent_count=[yes,no]
frequent_item=max(frequent_count)
if frequent_item==yes:
    replace=1
elif frequent_item==no:
    replace=2
for i in range(0,len(df_check)):
    if math.isnan(int(df_check.loc[i,"Loan_Status"])):
        df_check.loc[i,"Loan_Status"]=replace
```
